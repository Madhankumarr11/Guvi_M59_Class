```python
# Calculator

def calculator(a,b,operator):
  if operator=='+':
    return a+b
  elif operator=='-':
    return a-b
  elif operator=='*':
    return a*b
  elif operator=='/':
    return a/b
  else:
    return 'Invalid Operator'
```

```python
calculator(25,30,'*') # Postional Arg
```

```python
calculator('*',25,30)
```

```python
calculator(operator='+',a=25,b=30) # keyword Arg
```

```python
def Sqr_Mul(a,b):
  sqr_a=sqr(a)
  sqr_b=sqr(b)
  Mul_ab = (sqr_a*sqr_b)
  return Mul_ab
```
