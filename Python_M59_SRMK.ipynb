````python
#intersection()
#Definition: Rendu set-layum common-aa irukura items return pannum

A = {'Laptop', 'Mobile'}
B = {'Mobile', 'Tablet'}

print(A.intersection(B))
````

````python
#isdisjoint()
#Definition: Rendu set-kum common item illa-na True return pannum

A = {'Laptop'}
B = {'Mobile'}

print(A.isdisjoint(B))
````

````python
#issubset()
#Definition: Oru set, innoru set-oda part-aa irukka-nu check pannum

A = {'Mobile'}
B = {'Laptop', 'Mobile'}

print(A.issubset(B))
````

````python
#issuperset()
#Definition: Oru set, innoru set-a full-aa contain pannudha-nu check pannum

A = {'Laptop', 'Mobile'}
B = {'Mobile'}

print(A.issuperset(B))
````

````python
#symmetric_difference()
#Definition: Common items remove pannitu, unique items return pannum

A = {'Laptop', 'Mobile'}
B = {'Mobile', 'Tablet'}

print(A.symmetric_difference(B))
````

````python
#symmetric_difference_update()
#Definition: Symmetric difference result-a original set-la update pannum

A = {'Laptop', 'Mobile'}
B = {'Mobile', 'Tablet'}

A.symmetric_difference_update(B)
print(A)
````

#11.Class and Objects (13-Dec-2025)

**OOPS:**
1. Class and Objects [Constructor __init__, Attributes, Methods]
2. Inheritance
3. Polymorphism
4. Encapsulation
5. Abstraction

| **OOPS Concept**  | **Simple Meaning**                      | **Purpose / Why Used**              | **Syntax / Code Pattern**    | **Short Example**             | **Output / Result**        |
| ----------------- | --------------------------------------- | ----------------------------------- | ---------------------------- | ----------------------------- | -------------------------- |
| **Class**         | Blueprint / template                    | Group data + functions in one place | `python\nclass A:\n    pass` | `python\nclass Car: pass`     | Defines structure          |
| **Object**        | Instance of class                       | Create usable items from class      | `python\nobj = A()`          | `python\nc = Car()`           | Creates object             |
| **Constructor**   | Auto method when object created         | Initialize values                   | `python\n__init__(self)`     | `python\nself.name = n`       | Stores data in object      |
| **Attributes**    | Variables inside object                 | Hold object data                    | `python\nself.age`           | Name, Age, City etc.          | Data stored in object      |
| **Methods**       | Functions inside class                  | Object behavior                     | `python\ndef func(self):`    | `car.start()`                 | Runs behavior              |
| **Encapsulation** | Hide data (private variables)           | Protect data                        | `python\nself.__balance`     | Private balance               | Prevents direct access     |
| **Inheritance**   | One class gets features from another    | Reuse code                          | `python\nclass B(A):`        | `Dog(Animal)`                 | Child gets parent features |
| **Polymorphism**  | Same function name → different behavior | Flexibility                         | `python\nobj.func()`         | `car.horn()` vs `bike.horn()` | Different outputs          |
| **Abstraction**   | Hide complexity                         | Show only required functions        | `python\nclass A(ABC):`      | Payment → UPI                 | Clean, simple interface    |


````python
# Syntax for Class

class ClassName:

  def __init__(self,value1,value2): # init Method to Initaialize the object
    self.attribute1 = value1 # Attribues
    self.attribute2 = value2 # Attributes


  def Method1(self): # Method
    pass
  def Method2(self): # Method
    pass
````

````python
# 1. Class-ஐ உருவாக்குதல் (வரைபடம்)
class Car:
    # Constructor method - Object-ஐ ஆரம்பிக்கிறது
    def __init__(self, brand, color):
        self.brand = brand  # Attribute (பண்பு)
        self.color = color  # Attribute (பண்பு)

    # ஒரு Method (செயல்பாடு)
    def start_engine(self):
        return f"{self.color} நிற {self.brand} காரின் இன்ஜின் இயங்குகிறது."

# 2. Objects-ஐ உருவாக்குதல் (Class-லிருந்து உருவாக்கப்படும் உண்மையான கார்கள்)
car1 = Car("Toyota", "Red")
car2 = Car("Honda", "Blue")

# 3. Objects-ஐ பயன்படுத்துதல்
print(car1.brand)          # வெளியீடு: Toyota
print(car2.start_engine()) # வெளியீடு: Blue நிற Honda காரின் இன்ஜின் இயங்குகிறது.
````

|                     | `"w+"`             | Write + Read                | Overwrite then read          | `open("a.txt","w+")`       |

````python
class Swiggy_Analysis_Json(Json_File_Handling):
  def __init__(self,file_name):
    super().__init__(file_name)
    self.file_name=file_name
    self.data=self.read_file()['swiggy_orders']

  def Show_All_restaurant(self):
    return list(set([item['restaurant'] for item in self.data]))

  def restaurant_wise_order_counts(self):
    restaurant_order_counts={}
    for item in self.data:
      if item['status']=='delivered':
        restaurant=item['restaurant']
        if restaurant in restaurant_order_counts:
          restaurant_order_counts[restaurant]+=1
        else:
          restaurant_order_counts[restaurant]=1
    return restaurant_order_counts

  def restaurant_wise_order_revenue(self):
    restaurant_order_revenue={}
    for item in self.data:
      if item['status']=='delivered':
        restaurant=item['restaurant']
        if restaurant in restaurant_order_revenue:
          restaurant_order_revenue[restaurant]+=item['order_value']
        else:
          restaurant_order_revenue[restaurant]=item['order_value']

    return dict(sorted(restaurant_order_revenue.items(),key=lambda x:x[1],reverse=True))


  def Save_Report_Analysis(self,file_name):
    Data={'List_of_Restaurents':self.Show_All_restaurant(),
          'restaurant_wise_order_counts':self.restaurant_wise_order_counts(),
          'restaurant_wise_order_revenue':self.restaurant_wise_order_revenue()}

    json.dump(Data,open(file_name,'w'),indent=4)
    print("Report Saved Successfully")






````

#14.Algorithmic thinking with Python (21-12-2025)



###  **Definition**

> Algorithmic thinking is a *structured problem-solving approach* that involves analyzing a problem, identifying its components, and creating a sequence of well-defined steps (an algorithm) to solve it.

---

###  **Key Components of Algorithmic Thinking**

| Concept                 | Description                                                       | Example                                                                          |
| ----------------------- | ----------------------------------------------------------------- | -------------------------------------------------------------------------------- |
| **Decomposition**       | Breaking a large problem into smaller, manageable parts.          | Splitting “build an e-commerce app” into login, cart, and payment modules.       |
| **Pattern Recognition** | Identifying similarities or trends within problems.               | Recognizing that bubble sort and insertion sort both involve comparing elements. |
| **Abstraction**         | Focusing only on important details and ignoring unnecessary ones. | Ignoring the brand of a car when calculating fuel efficiency.                    |
| **Algorithm Design**    | Creating step-by-step instructions to solve the problem.          | Writing a flowchart or pseudocode for finding the largest number in a list.      |

---

###  **Example: Real-world Problem**

**Problem:** You need to make a cup of tea.

**Algorithmic Steps:**

1. Boil water.
2. Add tea leaves to the boiling water.
3. Let it brew for 2–3 minutes.
4. Add milk and sugar.
5. Strain the tea into a cup.
6. Serve hot.

This is algorithmic thinking — breaking down a task into sequential, logical steps.

---

###  **Example: Coding Problem**

**Problem:** Find the largest number in a list.

**Algorithm (in simple steps):**

1. Assume the first number is the largest.

1. O(1) - Constant time complexity means that the algorithm's runtime does not depend on the size of the input data. It performs the same number of operations regardless of the input size.

2. O(n) - Linear time complexity means that the time taken by the algorithm increases linearly with the size of the input. As the size of the input increases, the number of operations also increases linearly.

3. O(n^2) - Quadratic time complexity means that the time taken by the algorithm increases quadratically with the size of the input. Each increase in the input size results in a quadratic increase in the number of operations.
4. O(log(n)) - Logarithmic time complexity means that the time taken by the algorithm increases logarithmically with the size of the input. These algorithms typically divide the input in half at each step.

5. O(2^n) - Exponential time complexity means that the time taken by the algorithm doubles with each addition to the input data set.

````python
# Binary Search

def Binary_Search(arr,target):
  lower_bound=0                       #0-->4--->4
  upper_bound=len(arr)-1              #7---7--->4

  while lower_bound<=upper_bound:
    mid_index=(upper_bound+lower_bound)//2  #(0,7) #3 (4+7)//2-->5,(4+4)//2=4
    mid_element=arr[mid_index]                #25  -------->55-------->45

    if mid_element==target:          #(25,45) (55,45)(45,45)
      return f'{target} found at the index {mid_index}'
    elif mid_element<target:     #(25,45)(55,45)
      lower_bound=mid_index+1    #3+1=4
    else:
      upper_bound=mid_index-1    # 5-1=4

  return f'{target} not found in array'
````

````python
def GCD_Euclids(a, b):
    while b != 0:
        print('Before', 'A:', a, 'B:', b)
        # a-வை b-யாகவும், b-யை மீதியாகவும் (remainder) மாற்றுகிறோம்
        a, b = b, a % b
        print('After ', 'A:', a, 'B:', b)
        print("-" * 20)
    return a

# பங்க்ஷனுக்கு வெளியே இருந்து அழைக்கவும்
result = GCD_Euclids(1200, 2010)
print('மிகப்பெரிய பொது வகுத்தி (GCD):', result)
````

Real-world example:

2. Undo feature in software (Ctrl+Z): The most recent action is undone first.

3. Backtracking in maze-solving or games: Last path tried is the first one to be rolled back.

````python
class STACK: # FILO/ LIFO
  def __init__(self):
    self.__stack=[]

  def push(self, data):
    self.__stack.append(data)

  def pop(self):
    if len(self.__stack)!=0:
      return self.__stack.pop()
    else:
      print('Stack is Empty')


  def peek(self):
    if len(self.__stack)!=0:
      return self.__stack[-1]
    else:
      print('Stack is Empty')

  def is_empty(self):
    return len(self.__stack)==0

  def size(self):
    return len(self.__stack)

  def display(self):
    return f"Stack:{self.__stack}"

````

## Queue

![](https://cdn.programiz.com/sites/tutorial2program/files/queue.png)

Real-world example:

1. Line at a ticket counter: The first person in line is served first.

2. Print queue: Documents are printed in the order they are submitted.

3. Customer service call center: Calls are answered in the order they arrive.

````python
# FIFO/ LILO
class QUEUE:
  def __init__(self):
    self.queue=[]

  def enqueue(self,data):
    """
    This Function will Help us to Add Data into Queue
    """
    self.queue.append(data)
    print(f'Enqueued:{data}')

  def dequeue(self):
    if len(self.queue)!=0:
      return self.queue.pop(0)
    else:
      return 'Queue is Empty'

  def front(self):
    if len(self.queue)!=0:
      return self.queue[0]
    else:
      return 'Queue is Empty'

  def rear(self):
    if len(self.queue)!=0:
      return self.queue[-1]
    else:
      return 'Queue is Empty'

  def is_empty(self):
    return len(self.queue)==0

  def size(self):
    return len(self.queue)
````

## Binary Search Tree (BST)

A Binary Search Tree (BST) is a node-based binary tree data structure that has the following properties:

*   The left subtree of a node contains only nodes with keys less than the node's key.
*   The right subtree of a node contains only nodes with keys greater than the node's key.
*   The left and right subtrees each must also be a binary search tree.
*   There must be no duplicate nodes.

These properties allow for efficient searching, insertion, and deletion of elements. The time complexity of these operations is typically O(log n) in the average case, where n is the number of nodes in the tree. However, in the worst case (a skewed tree), the time complexity can degrade to O(n).

Common operations on a BST include:

*   **Search:** Finding a specific node in the tree.
*   **Insertion:** Adding a new node to the tree while maintaining the BST properties.
*   **Deletion:** Removing a node from the tree while maintaining the BST properties.
*   **Traversal:** Visiting all nodes in the tree in a specific order (e.g., in-order, pre-order, post-order).

BSTs are widely used in various applications, such as implementing dictionaries, sets, and symbol tables.

````python
class TreeNode:
  def __init__(self,data=None):
    self.data=data
    self.left=None
    self.right=None
````

````python
# Insert into BST

def insert(root, data):
  if root is None:
    return TreeNode(data)
  if data<root.data:
    root.left = insert(root.left, data)
  else:
    root.right = insert(root.right, data)

  return root
````

````python
root=None
L=[20,30,40,50,55,25,110,250,600,15]

for value in L:
  root=insert(root,value)
````

````python
# Inorder Traversel (Left-->RootNode-->Right)
def InOrderTraversel(node):
  if node is None:
    return
  InOrderTraversel(node.left)
  print(node.data,end=', ')
  InOrderTraversel(node.right)

InOrderTraversel(root)
````

````python
# PreOrder Traversel(RootNode-->Left-->Right)

def PreOrderTraversel(node):
  if node is None:
    return
  print(node.data,end=', ')
  PreOrderTraversel(node.left)
  PreOrderTraversel(node.right)

PreOrderTraversel(root)
````

````python
# PostOrder Traversel(Left-->Right---RootNode)

def PostOrderTraversel(node):
  if node is None:
    return

  PostOrderTraversel(node.left)
  PostOrderTraversel(node.right)
  print(node.data,end=', ')

PostOrderTraversel(root)
````

````python
def search(node,target):
  if node is None:
    return None
  if node.data==target:
    return True
  elif target<node.data:
    return search(node.left,target)
  else:
    return search(node.right,target)

````

```sql
        AVG(total_amount) AS avg_order_value
    FROM fact_orders
    GROUP BY restaurant_id
),
overall_avg AS (
    SELECT AVG(avg_order_value) AS overall_avg_value
    FROM restaurant_avg
)
SELECT
    r.restaurant_id,
    d.restaurant_name,
    r.avg_order_value
FROM restaurant_avg r
JOIN dim_restaurant d
    ON r.restaurant_id = d.restaurant_id,
     overall_avg o
WHERE r.avg_order_value > o.overall_avg_value
ORDER BY r.avg_order_value DESC;
```

## 13. Customer Revenue Tier Classification

```sql
WITH customer_spend AS (
    SELECT
        c.customer_id,
        c.customer_name,
        SUM(o.total_amount) AS total_spent
    FROM dim_customer c
    JOIN fact_orders o
        ON c.customer_id = o.customer_id
    GROUP BY c.customer_id, c.customer_name
)
SELECT
    customer_id,
    customer_name,
    total_spent,
    CASE
        WHEN total_spent > 5000 THEN 'Platinum'
        WHEN total_spent BETWEEN 2500 AND 5000 THEN 'Gold'
        WHEN total_spent BETWEEN 1000 AND 2500 THEN 'Silver'
        ELSE 'Bronze'
    END AS customer_tier
FROM customer_spend
ORDER BY total_spent DESC;
```

## 14. Restaurant Menu Breadth vs Revenue

```sql
WITH item_counts AS (
    SELECT
        restaurant_id,
        COUNT(*) AS number_of_items
    FROM dim_item
    GROUP BY restaurant_id
),
restaurant_revenue AS (
    SELECT
        restaurant_id,
        SUM(total_amount) AS total_revenue
    FROM fact_orders
    GROUP BY restaurant_id
)
SELECT
    r.restaurant_id,
    d.restaurant_name,
    COALESCE(i.number_of_items, 0) AS number_of_items,
    COALESCE(rr.total_revenue, 0) AS total_revenue
FROM dim_restaurant d
LEFT JOIN item_counts i
    ON d.restaurant_id = i.restaurant_id
LEFT JOIN restaurant_revenue rr
    ON d.restaurant_id = rr.restaurant_id
ORDER BY total_revenue DESC, number_of_items DESC;
```

## 15. Order Leakage – Inactive Customers (No Orders in Last 45 Days)

```sql
WITH last_order AS (
    SELECT
        customer_id,
        MAX(order_timestamp) AS last_order_timestamp
    FROM fact_orders
    GROUP BY customer_id
)
SELECT
    c.customer_id,
    c.customer_name,
    l.last_order_timestamp
FROM dim_customer c
LEFT JOIN last_order l
    ON c.customer_id = l.customer_id
WHERE l.last_order_timestamp IS NULL
   OR l.last_order_timestamp < NOW() - INTERVAL '45 days'
ORDER BY last_order_timestamp NULLS FIRST, c.customer_id;
```
